# Example 13

```python
import numpy as np

# Define a matrix
A = np.array([[4, 7], [2, 6]])

# Calculate the determinant
determinant = np.linalg.det(A)

# Calculate the inverse if the matrix is invertible
if determinant != 0:
    inverse_A = np.linalg.inv(A)
    print("Inverse of A:\n", inverse_A)
else:
    print("Matrix A is singular and has no inverse.")

print("Determinant of A:", determinant)
```

```python
import numpy as np

# Define a square matrix
A = np.array([[2, 1], [1, 3]])

# Raise matrix to the third power
A_cubed = np.linalg.matrix_power(A, 3)

print("A cubed:\n", A_cubed)
```

```python
import numpy as np
from scipy.linalg import expm

# Define a square matrix
A = np.array([[0, 1], [-1, 0]])

# Calculate the matrix exponential
expA = expm(A)

print("Matrix exponential of A:\n", expA)
```
