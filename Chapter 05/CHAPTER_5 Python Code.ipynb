# Example 5.14

```python
def arithmetic_operations(a, b):
    sum_result = a + b
    difference = a - b
    return sum_result, difference

s, d = arithmetic_operations(10, 5)
print(f"Sum: {s}, Difference: {d}")
```

# Example 5.17

```python
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n - 1)

print(factorial(5))  # 120
```
